## Fiesta Case Study Analysis

Fiesta, an online retailer, was a small and lean firm with 20 employees. In the absence of data to forecast purchasing trends, the startup was operating on a risk averse, pull-based business model: with each order that came through the website, a procurement 
plan was drawn up. Items would then be sourced from a network of suppliers and shipped to a designated DC, before making their way to the customer.

The purpose of this notebook is the to come up with strategies and inferences that can help aid in the decision making with regard to the supply chain sector of the company.

```python
skucode_df = fiesta.select('SKUCode','Quantity','TotalPrice')\
                    .groupby('SKUCode').sum().orderBy(desc('sum(TotalPrice)'))\
                    .drop('sum(SKUCode)').na.drop().toPandas().set_index('SKUCode')
skucode_df.columns = ['Quantity','TotalPrice']
skucode_df.head()
```

```python
skucode_df.count()
```

```python
skucode_df['TotPrice%'] = skucode_df['TotalPrice'].apply(lambda x : x*100/skucode_df['TotalPrice'].sum())
skucode_df['CumPrice%'] = skucode_df['TotPrice%'].cumsum()

skucode_df['TotQuantity%'] = skucode_df['Quantity'].apply(lambda x : x*100/skucode_df['Quantity'].sum())
skucode_df['CumTotQuantity%'] = skucode_df['TotQuantity%'].cumsum()

skucode_df = skucode_df[['Quantity','TotQuantity%','CumTotQuantity%','TotalPrice','TotPrice%','CumPrice%']]
skucode_df
```

## Getting list of unique DC, Supplier and Customer cities

```python
cities_dc =  fiesta.select('DCCity').distinct().rdd.flatMap(lambda x :x ).collect()
cities_sup = fiesta.select('SuppCity').distinct().rdd.flatMap(lambda x :x ).collect()
cities_cus = fiesta.select('CustCity').distinct().rdd.flatMap(lambda x :x ).collect()
```

## Obtaining geolocations of the above cities

```python
#May take approximatley 5-10 mins to obtain co-ordinates
def Return_Lat_Long_Cities(cities):
    geolocator = Nominatim(user_agent="Assignment")
    lat_lon = []
    for city in cities: 
        try:
            location = geolocator.geocode(city)
            if location:
                lat_lon.append(location)
        except GeocoderTimedOut as e:
            return("Error: geocode failed on input %s with message %s"%(city, e))
    return lat_lon

lat_lon_dc = Return_Lat_Long_Cities(cities_dc)
lat_lon_sup = Return_Lat_Long_Cities(cities_sup)
lat_lon_cus = Return_Lat_Long_Cities(cities_cus)
```

```python
# lat_lon_sup[:5]
```

```python
df_dc = pd.DataFrame(lat_lon_dc, columns=['City Name', 'Coordinates'])
geometry_dc = [Point(x[1], x[0]) for x in df_dc['Coordinates']]

df_sup = pd.DataFrame(lat_lon_sup, columns=['City Name', 'Coordinates'])
geometry_sup = [Point(x[1], x[0]) for x in df_sup['Coordinates']]

df_cus = pd.DataFrame(lat_lon_cus, columns=['City Name', 'Coordinates'])
geometry_cus = [Point(x[1], x[0]) for x in df_cus['Coordinates']]
```

### Creating a GeoDataframe

```python
crs = {'init': 'epsg:3395'}

geo_df_dc = gpd.GeoDataFrame(df_dc, crs=crs, geometry=geometry_dc)
geo_df_sup = gpd.GeoDataFrame(df_sup, crs=crs, geometry=geometry_sup)
geo_df_cus = gpd.GeoDataFrame(df_cus, crs=crs, geometry=geometry_cus)
geo_df_sup.head()
```

```python
geo_df_dc['City Name'] = geo_df_dc['City Name'].apply(lambda x:x.split(",")[0])
geo_df_sup['City Name'] = geo_df_sup['City Name'].apply(lambda x:x.split(",")[0])
geo_df_cus['City Name'] = geo_df_cus['City Name'].apply(lambda x:x.split(",")[0])
```
